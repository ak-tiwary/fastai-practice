## Jaguar vs Cheetah Classifier

Add an image of a jaguar or a cheetah and the model will predict whether it is a jaguar or a cheetah with 90% accuracy.

```python
path = Path('./gdrive/MyDrive')
learn_inf = load_learner(path/'export.pkl',cpu=True)
```

```python
btn_upload = widgets.FileUpload()
out_pl = widgets.Output()
lbl_pred = widgets.Label()
```

```python
def on_data_change(change):
  lbl_pred.value = ''
  img = PILImage.create(btn_upload.data[-1])
  out_pl.clear_output()
  with out_pl: display(img.to_thumb(256,256))
  pred, pred_idx, prob = learn_inf.predict(img)
  lbl_pred.value = f'Prediction: {pred}; Probability: {prob[pred_idx]:.04f}'
```

```python
btn_upload.observe(on_data_change,names=['data'])
```

```python
display(VBox([widgets.Label('Select the big cat'), btn_upload, out_pl,lbl_pred]))
```
